# Hypothesis Testing

```python
dfc.columns
```

```python
dfc
```

```python
df2020=dfc[dfc['Date'].dt.year==2020]
df2021=dfc[dfc['Date'].dt.year!=2020]

df2020nc=df2020['New Cases']
df2021nc=df2021['New Cases']
```

```python
df2020nc
```

```python
df2021nc
```

Assumption : New Cases on 2021 is lower than new cases in 2020 --> Sample vs Sample Dependent (two paired test)

- H0 = Mean 2021 Covid New Cases is less than or equeal to than the mean of 2020 Covid New Cases
- H1 = Mean 2021 Covid New Cases is higher than the mean of 2020 Covid New Cases
- Critical Value is 0,05

```python
df2020mean = df2020nc.mean()
df2021mean = df2021nc.mean()
```

```python
print(df2020mean)
print(df2021mean)
```

```python
t_stat,p_val = stats.ttest_rel(df2020nc.sample(7000),df2021nc.sample(7000)) #we take 40000 sample of each data to make the size the same.
print('P-value:',p_val)
```
